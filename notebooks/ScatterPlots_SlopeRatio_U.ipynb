# Scatter plots for various coefficients (slope ratio, Ux, Uy, and dU/dy)

```python
#!/usr/bin/env python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# increase dpi and use latex rendering
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 160
mpl.rc('text', usetex=True)

# Set macro for paths
run = "../run/"
```

```python
UInf = 69.4
nuInf = 1.388e-05

model_sa = "SpalartAllmaras"
yp = ['30', '50', '100']
res_list = ['1e-5']
no_correction_iter = '50'
final_iter = '100'
```

```python
fields_path = [run + 'turbulentFlatPlate_wmSF_count50_100iter/yplus_{}/{}yp_surfaceFields_iter_{}.csv'.format(yp[j], yp[j], final_iter) for j in range(len(yp))]
fields_data = [pd.read_csv(fields_path[i + len(res_list)*j]) for j in range(len(yp)) for i in range(len(res_list))]
fields_data[0].head()
```

```python
fields_path_x0 = [run + 'turbulentFlatPlate_wmSF_count50_100iter_x0/yplus_{}/{}yp_surfaceFields_iter_{}.csv'.format(yp[j], yp[j], final_iter) for j in range(len(yp))]
fields_data_x0 = [pd.read_csv(fields_path_x0[i + len(res_list)*j]) for j in range(len(yp)) for i in range(len(res_list))]
fields_data_x0[0].head()
```

```python
x = fields_data[0]['ccx']
```

## Slope ratio, Ux, and Uy at the first cell face (100 iteration step, after correction at the face)

```python
lw = 1.0
for j in range(len(yp)):
    fig, ax = plt.subplots(1, 1, figsize=(8, 3))    
    plt.grid()
    ax.set_xlabel(r"$x$")    
    plt.title(r"Slope Ratio for y+ = {} at 100 iteration step (after correction)".format(yp[j]))    
    ax.scatter(x, fields_data[0 + len(res_list)*j]['Slope_ratio'], label = "{} y+ = {}, started from x = -1".format(model_sa, yp[j]), c = "C{:d}".format(0), lw = lw, marker = ".")
    ax.scatter(x, fields_data_x0[0 + len(res_list)*j]['Slope_ratio'], label = "{} y+ = {}, started from x = 0".format(model_sa, yp[j]), c = "C{:d}".format(1), lw = lw, marker = ".")
    ax.legend(fontsize = 8)
```

```python
lw2 = 1.0
for j in range(len(yp)):
    fig2, ax2 = plt.subplots(1, 1, figsize=(8, 3))    
    plt.grid()
    ax2.set_xlabel(r"$x$")    
    plt.title(r"Ux 1st Cell Center for y+ = {} at 100 iteration step (after correction)".format(yp[j]))    
    ax2.scatter(x, fields_data[0 + len(res_list)*j]['Ux_1st_cell'], label = "{} y+ = {}, started from x = -1".format(model_sa, yp[j]), c = "C{:d}".format(0), lw = lw2, marker = ".")
    ax2.scatter(x, fields_data_x0[0 + len(res_list)*j]['Ux_1st_cell'], label = "{} y+ = {}, started from x = 0".format(model_sa, yp[j]), c = "C{:d}".format(1), lw = lw2, marker = ".")
    ax2.legend(fontsize = 8)
```

```python
lw3 = 1.0
for j in range(len(yp)):
    fig3, ax3 = plt.subplots(1, 1, figsize=(8, 3))    
    plt.grid()
    ax3.set_xlabel(r"$x$")    
    plt.title(r"Uy 1st Cell Center for y+ = {} at 100 iteration step (after correction)".format(yp[j]))
    ax3.scatter(x, fields_data[0 + len(res_list)*j]['Uy_1st_cell'], label = "{} y+ = {}, started from x = -1".format(model_sa, yp[j]), c = "C{:d}".format(0), lw = lw3, marker = ".")
    ax3.scatter(x, fields_data_x0[0 + len(res_list)*j]['Uy_1st_cell'], label = "{} y+ = {}, started from x = 0".format(model_sa, yp[j]), c = "C{:d}".format(1), lw = lw3, marker = ".")
    ax3.legend(fontsize = 8)
```
