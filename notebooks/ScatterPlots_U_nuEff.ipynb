# Scatter plots for various coefficients

```python
#!/usr/bin/env python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# increase dpi and use latex rendering
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 160
mpl.rc('text', usetex=True)

# Set macro for paths
run = "../../run/"
```

```python
UInf = 69.4
nuInf = 1.388e-05

model_sa = "SpalartAllmaras"
#yp = ['0.05', '1', '2', '5', '10', '30', '50', '100']
yp = ['100']
res_list = ['1e-5']
iter_list = [i+1 for i in range(100)]
```

```python
fields_path = [run + 'turbulentFlatPlate_wmSF_1stface_U0.9_sigmoid/yplus_{}/{}yp_surfaceFields_iter_{}.csv'.format(yp[j], yp[j], iter_list[i]) for j in range(len(yp)) for i in range(len(iter_list))]
fields_data = [pd.read_csv(fields_path[i + len(res_list)*j]) for j in range(len(yp)) for i in range(len(iter_list))]
fields_data[0].head()
```

```python
x = fields_data[0]['ccx']
```

```python
#style = ["-", "--", "-.", ":"]
lw = 1.0
for j in range(len(yp)):
    fig, ax = plt.subplots(1, 1, figsize=(8, 3))    
    plt.grid()
    ax.set_xlabel(r"$x$")    
    plt.title(r"Velocity Ux for y+ = {}".format(yp[j]))    
    for i in range(10):
        ax.scatter(x, fields_data_50[i + len(res_list)*j]['Ux_1st_cell'], label = "{} y+ = {} iteration {}".format(model_sa, yp[j], iter_list[i]), c = "C{:d}".format(i), lw = lw, marker = ".")
        ax.legend(fontsize = 8, bbox_to_anchor=(1.05, 0.5))
```
